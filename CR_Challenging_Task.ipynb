```python
import requests  # Send request
from bs4 import BeautifulSoup  # Parse HTML
import csv  # CSV handling
import matplotlib.pyplot as plt  # Plotting (graph)

def simple_scraper():
    # Get page
    page = requests.get("http://quotes.toscrape.com")
    soup = BeautifulSoup(page.text, 'html.parser')

    quotes = []
    authors = {}

    # Find quotes
    for q in soup.find_all('div', class_='quote'):
        quote = q.find('span', class_='text').text.strip('‚Äú‚Äù')  # Quote text
        author = q.find('small', class_='author').text          # Author name
        tags = ", ".join(tag.text for tag in q.find_all('a', class_='tag'))  # Tags

        quotes.append([quote, author, tags])  # Save quote
        authors[author] = authors.get(author, 0) + 1  # Count author

        # Print quote, author, and tags
        print("Quote:", quote)
        print("Author:", author)
        print("Tags:", tags)
        print("-" * 50)

    # Save CSV
    with open('quotes.csv', 'w', newline='', encoding='utf-8') as f:
        csv.writer(f).writerows([['Quote', 'Author', 'Tags']] + quotes)

    print("Saved in quotes.csv ‚úÖ")

    # Show graph
    plt.bar(authors.keys(), authors.values())
    plt.xticks(rotation=45)
    plt.title("Quotes per Author")
    plt.tight_layout()
    plt.show()

# Run function
simple_scraper()
```
